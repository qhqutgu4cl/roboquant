```kotlin
%use roboquant@url[https://roboquant.org/roboquant.json](1.1.0)
```

```kotlin
val timeline = Timeframe.fromYears(2019, 2020).toTimeline(1.days)
val feed = RandomWalkFeed(timeline)
```

```kotlin
val strategy = EMAStrategy()
val roboquant =  Roboquant(strategy, AccountMetric())
val periods = feed.split(1.years)
periods.forEach { 
    roboquant.run(feed, it)   
}
```

```kotlin
val data = roboquant.logger.getMetric("account.equity").takeLast(10)
MetricChart(data)
```

```kotlin
roboquant.reset()

feed.sample(100, 5).forEach {
    roboquant.run(feed, it) 
}
```

```kotlin
val data = roboquant.logger.getMetric("account.equity")
MetricChart(data, useTime=false)
```
